```python
#Load the libraries
import pandas as pd
import re
!pip install nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load dataset
data = pd.read_csv("23k_r_gaming_comments_sentiments.csv")
```

```python
# Display the first few rows
data.head()
```

```python
# Data Cleaning: Removing URLs, special characters, and unnecessary whitespace
data['Cleaned_Comment'] = data['Comment'].apply(lambda x: re.sub(r'http\S+', '', x))  # Remove URLs
data['Cleaned_Comment'] = data['Cleaned_Comment'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))  # Remove special characters
data['Cleaned_Comment'] = data['Cleaned_Comment'].apply(lambda x: x.lower())  # Convert to lowercase

# Tokenization
data['Tokens'] = data['Cleaned_Comment'].apply(word_tokenize)

# Stopword Removal
stop_words = set(stopwords.words('english'))
data['Tokens'] = data['Tokens'].apply(lambda x: [word for word in x if word not in stop_words])

# Display the first few rows after processing
data[['Comment', 'Cleaned_Comment', 'Tokens']].head()
```

```python
# Using Python's string methods for tokenization, stopword removal, and stemming

# Tokenization using string split method
data['Tokens'] = data['Cleaned_Comment'].apply(lambda x: x.split())

# Basic list of stopwords
basic_stopwords = [
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're", "you've", "you'll", 
    "you'd", 'your', 'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she', "she's", 
    'her', 'hers', 'herself', 'it', "it's", 'its', 'itself', 'they', 'them', 'their', 'theirs', 'themselves', 
    'what', 'which', 'who', 'whom', 'this', 'that', "that'll", 'these', 'those', 'am', 'is', 'are', 'was', 
    'were', 'be', 'been', 'being', 'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an', 
    'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of', 'at', 'by', 'for', 'with', 
    'about', 'against', 'between', 'into', 'through', 'during', 'before', 'after', 'above', 'below', 'to', 
    'from', 'up', 'down', 'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then', 'once'
]

# Stopword Removal
data['Tokens'] = data['Tokens'].apply(lambda x: [word for word in x if word not in basic_stopwords])

# Display the first few rows after processing
data[['Comment', 'Cleaned_Comment', 'Tokens']].head()
```

```python
# Flatten the list of tokens and compute the frequency of each token
all_tokens = [token for tokens in data['Tokens'] for token in tokens]
token_frequency = Counter(all_tokens)

# Get the top 20 most common words
top_words = token_frequency.most_common(20)

top_words
```

```python
# Extracting comments related to the theme "Game Mechanics and Experience"
relevant_keywords = ["game", "play", "controller", "games"]
relevant_comments = data[data['Tokens'].apply(lambda tokens: any(word in tokens for word in relevant_keywords))]

# Sampling a subset of these comments for qualitative analysis
sample_comments = relevant_comments.sample(10, random_state=42)

sample_comments[['Comment', 'sentiment']]
```

```python
# Extracting comments related to the theme "User Feelings and Opinions"
relevant_keywords_feelings = ["like", "dont", "just", "so", "not"]
relevant_comments_feelings = data[data['Tokens'].apply(lambda tokens: any(word in tokens for word in relevant_keywords_feelings))]

# Sampling a subset of these comments for qualitative analysis
sample_comments_feelings = relevant_comments_feelings.sample(10, random_state=42)

sample_comments_feelings[['Comment', 'sentiment']]
```

```python
# Extracting comments related to the theme "Community and Interactions"
relevant_keywords_community = ["people"]
relevant_comments_community = data[data['Tokens'].apply(lambda tokens: any(word in tokens for word in relevant_keywords_community))]

# Sampling a subset of these comments for qualitative analysis
sample_comments_community = relevant_comments_community.sample(10, random_state=42)

sample_comments_community[['Comment', 'sentiment']]
```

```python
# Extracting longer comments as they are more likely to contain detailed narratives
long_comments = data[data['Cleaned_Comment'].str.split().str.len() > 50]

# Sampling a subset of these long comments for narrative analysis
sample_long_comments = long_comments.sample(5, random_state=42)

sample_long_comments[['Comment', 'sentiment']]
```
